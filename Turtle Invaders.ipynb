```python
import turtle
```

```python
defender = turtle
toleft = 0
toright = 0

defender.penup()
defender.shape('square')
defender.lt(180)
defender.goto(0,-316)

def left():
    defender.fd(10)
    toleft += 1
    toright -= 1

def right():
    defender.bk(10)
    toleft -= 1
    toright += 1

if toright <= -34 and toleft >= 34:
    defender.onkey(right, "Right")
    defender.onkey(right, "d")
    defender.listen()
elif toright >= 34 and toleft <= -34:
    defender.onkey(left, "Left")
    defender.onkey(left, "a")
    defender.listen()

else:
    defender.onkey(left, "Left")
    defender.onkey(left, "a")
    defender.onkey(right, "Right")
    defender.onkey(right, "d")
    defender.listen()

turtle.mainloop()
```
