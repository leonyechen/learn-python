```python
def Leons_slot():
    import random
    a = random.randint(1, 9)
    b = random.randint(1, 9)
    c = random.randint(1, 9)
    output = (a,b,c)
    if a == b and b == c:
        print(output)
        print('YOU JUST WON €500!!!')
    else:
        print(output)
        print('YOU LOST, BETTER LUCK NEXT TIME')
    return output
```

```python
def Leons_multiplication(*args):
    output = None
    for num in args:
        if output is None:
            output = num
            continue
        output *= num
    return output
```

# Leons Easter Date
######            find out what date is Easter in ANY year

```python
special_list = list(map(int, input().split()))
y = special_list[0]
a = y % 19
b = y // 100
c = y % 100
d = b // 4
e = b % 4
f = (b + 8) // 25
g = (b - f + 1) // 3
h = (19 * a + b - d - g + 15) % 30
i = c // 4
k = c % 4
l = (32 + 2 * e + 2 * i - h - k) % 7
m = (a + 11 * h + 22 * l) // 451
month = (h + l - 7 * m + 114) // 31
date = ((h + l - 7 * m + 114) % 31) + 1
print(f'{date}/{month}/{y}')
```
