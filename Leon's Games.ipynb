# Leons_addition_quiz
##### now you can quiz yourself!!!

```python
def Leons_addition_quiz(range1, range2):
    import random
    i = 0
    num1 = None
    num2 = None
    real_answer = None
    ques_selecter = None
    answer = None
    answer1 = None
    answer2 = None
    answer3 = None
    answer4 = None
    real_answer2 = None
    answer_selecter = None
    right_answers = 0
    wrong_answers = 0
    ques_selection = None
    play_again = None
    while True:
        num1 = random.randint(range1, range2 // 2)
        num2 = random.randint(range1, range2 // 2)
        real_answer = num1 + num2
        ques_selecter = random.randint(1, 10)
        answer_selecter = random.randint(0, 40)
        if answer_selecter <= 10:
            answer1 = real_answer
            real_answer2 = 'a'
            answer2 = random.randint(real_answer - 25, real_answer + 25)
            answer3 = random.randint(real_answer - 25, real_answer + 25)
            answer4 = random.randint(real_answer - 25, real_answer + 25)
        elif answer_selecter > 10 and answer_selecter <= 20:
            real_answer2 = 'b'
            answer1 = random.randint(real_answer - 25, real_answer + 25)
            answer2 = real_answer
            answer3 = random.randint(real_answer - 25, real_answer + 25)
            answer4 = random.randint(real_answer - 25, real_answer + 25)
        elif answer_selecter > 20 and answer_selecter <= 30:
            real_answer2 = 'c'
            answer1 = random.randint(real_answer - 25, real_answer + 25)
            answer2 = random.randint(real_answer - 25, real_answer + 25)
            answer3 = real_answer
            answer4 = random.randint(real_answer - 25, real_answer + 25)
        elif answer_selecter > 30 and answer_selecter <= 40:
            real_answer2 = 'd'
            answer1 = random.randint(real_answer - 25, real_answer + 25)
            answer2 = random.randint(real_answer - 25, real_answer + 25)
            answer3 = random.randint(real_answer - 25, real_answer + 25)
            answer4 = real_answer
        if ques_selecter <= 5:
            ques_selection = True
        elif ques_selecter > 5:
            ques_selection = False
        if ques_selection:
            print(f'{num1} + {num2} = ?')
            print(f"{answer1}(a) {answer2}(b) {answer3}(c) {answer4}(d)")
            print()
            answer = input()
            print()
            if answer == real_answer2:
                right_answers += 1
                print('Great Job')
                print('')
            else:
                wrong_answers += 1
                print('Incorrect')
                print('')
        elif ques_selection is False:
            print(f'{num1} + {num2} = ?')
            print()
            answer = int(input())
            print()
            if answer == real_answer:
                right_answers += 1
                print('Brilliant')
                print('')
            else:
                wrong_answers += 1
                print('Wrong')
                print('')
        i += 1
        if i == 10:
            print("Play again? (yes/no)")
            print()
            play_again = input()
            print()
            if play_again == "yes":
                continue
            elif play_again == 'no':
                print(f'You answered {right_answers} right and {wrong_answers} wrong')
                break
            else:
                print("I will take that as a no")
                print(f'You answered {right_answers} right and {wrong_answers} wrong')
                break
```

# Turtle Race

##### Roll the die!!!

```python
import turtle
import random
import time
```

```python
turtle.title('Turtle Race')
player_one = turtle.Turtle()
player_one.color("green")
player_one.shape("turtle")
player_one.penup()
player_one.goto(-200,100)
player_two = player_one.clone()
player_two.color("blue")
player_two.penup()
player_two.goto(-200,-100)
player_one.goto(300,60)
player_one.pendown()
player_one.circle(40)
player_one.penup()
player_one.goto(-200,100)
player_two.goto(300,-140)
player_two.pendown()
player_two.circle(40)
player_two.penup()
player_two.goto(-200,-100)
die = [1,2,3,4,5,6]

player_one_name = input('Player one name:')
player_two_name = input('Player two name:')

while True:
    input('Press enter to roll the die')
    number = random.choice(die)
    print(f'You rolled {number}')
    player_one.fd(10*number)
    input('Press enter to roll the die')
    number = random.choice(die)
    print(f'You rolled {number}')
    player_two.fd(10*number)
    
    if player_one.pos() >= (300, 100):
        print(f'{player_one_name} Wins!')
        break
        
    if player_one.pos() >= (300, -100):
        print(f'{player_two_name} Wins!')
        break

while True:
    player_one.goto(300, 100)
    time.sleep(0.5)
    player_two.goto(300, -100)
```

```python
def Leons_Tank_Game():
    import time
    import colorama
    
    Tank = [colorama.Fore.LIGHTGREEN_EX + "     ░███████]▄▄▄▄▄▄▄▄▄▄▄▄ ",
            colorama.Fore.LIGHTGREEN_EX + "▂▄▅█████████▅▄▃▂"
            ,colorama.Fore.LIGHTGREEN_EX + "[███████████████████]     ",
            colorama.Fore.BLACK + "◥⊙▲⊙▲⊙▲⊙▲⊙▲⊙▲⊙◤   "]
    Wall = '||'
    wall_or_not = False
    distance = ''
    current_distance = ''
    bullet_distance = ''
    wall_distance = '                                 '
    
    while True:
        print('\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n')
        print('\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n')
        print('\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n')
        print('\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n')
        
        if wall_or_not == False:
            for line in Tank:
                print(current_distance + line)

            print('–––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––',end="")
            print('––––––––––––––––––––––––––––')
        
        else:
            for index,line in enumerate(Tank):
                if index >= 1:
                    print(current_distance + line + wall_distance + Wall)
                else:
                    print(current_distance + line)

            print('–––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––',end="")
            print('––––––––––––––––––––––––––––')

        choice = input()
        if choice == 'a' and len(current_distance) > 0:
            current_distance = current_distance[:-3:]
            wall_distance += '   '
        elif choice == 'd':
            current_distance += '   '
            wall_distance = wall_distance[:-3:]
        elif choice == 'wall':
            wall_or_not = True
        elif choice == 'quit':
            break
        if len(current_distance) == 87:
            current_distance = ''
```
